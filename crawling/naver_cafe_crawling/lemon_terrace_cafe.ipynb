```python
write_15 = driver.find_element(By.ID,'listSizeSelectDiv')
write_15.click()
time.sleep(1)

# 게시글 50개로 변경
write_50 = driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[2]/div[3]/ul/li[7]/a')
write_50.click()
# print(write_50)
```

```python
link = []
userid = []
date = []

page = 1

while True:

    
    
    try:
        
        try:
            driver.switch_to.frame("cafe_main")
        except:
            pass

        for i in range(1,12):
            
            driver.find_element(By.XPATH,f'//*[@id="main-area"]/div[7]/a[{i}+1]').click()

            html_source = driver.page_source
            soup_source = BeautifulSoup(html_source, 'html.parser')

            for j in range(50):

                link_data = soup_source.find_all("a",class_="article")[j]['href']
                username_data = soup_source.find_all("td",class_="td_name")[j].get_text().strip()
                date_data = soup_source.find_all("td",class_="td_date")[j].get_text()

                link.append(link_data)
                userid.append(username_data)
                date.append(date_data)

            print(page)
        
            if page == 41:
                break
                
            page += 1        
            time.sleep(3)

        if page == 41:
            break

        driver.find_element(By.XPATH, '//*[@id="main-area"]/div[7]/a[11]').click()

           
    except:
        # print('더이상 페이지가 존재하지 않음')
        pass
    
    
    
```

```python
print(len(link))
print(len(userid))
print(len(date))
```

```python
link_df = pd.DataFrame({'link':link,'userid':userid,'date':date})
link_df = link_df.drop_duplicates()
```
